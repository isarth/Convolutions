```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
```

```python
im=cv2.imread('tesla.png')
print im.shape
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
im=cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
print im.shape
#print im
plt.imshow(im,cmap='gray')
```

```python
kernel= np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]],dtype='float32')
print kernel
```

```python
def hadamard_prod(mat_01,mat_02):
    m,n= mat_01.shape
    p,q=mat_02.shape
            
    return mat_01*mat_02

def convolution2D(image, kernel):
    m,n= image.shape
    p,q= kernel.shape
    out= np.zeros((m-p+1, n-q+1))
    
    #convolve here
    for ix in range(out.shape[0]):
        for iy in range(out.shape[1]):
            im_patch= image[ix:ix+p, iy:iy+q]
            prod= hadamard_prod(im_patch, kernel)
            val= prod.sum()
            if val<0:
                val= 0.0
            elif val>255:
                val==255.0
            out[ix,iy]= val
    return out
```

```python
c= convolution2D(im, kernel)
print c.shape

plt.figure(0)
plt.axis("off")
plt.imshow(c,cmap="gray")
plt.show()
```
